```python
# --- 1. Calculate IQR bounds using the UN-SCALED data (df_encoded) ---

# Calculate bounds for Age
Q1_age = df_encoded['Age'].quantile(0.25)
Q3_age = df_encoded['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age
lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age

# Calculate bounds for Fare
Q1_fare = df_encoded['Fare'].quantile(0.25)
Q3_fare = df_encoded['Fare'].quantile(0.75)
IQR_fare = Q3_fare - Q1_fare
lower_bound_fare = Q1_fare - 1.5 * IQR_fare
upper_bound_fare = Q3_fare + 1.5 * IQR_fare

# --- 2. Create boolean mask (conditions) based on the UN-SCALED data ---

# Condition for rows within the Age IQR bounds
age_condition = (df_encoded['Age'] >= lower_bound_age) & (df_encoded['Age'] <= upper_bound_age)

# Condition for rows within the Fare IQR bounds
fare_condition = (df_encoded['Fare'] >= lower_bound_fare) & (df_encoded['Fare'] <= upper_bound_fare)

# Combine the conditions (a row must satisfy BOTH to be kept)
final_condition = age_condition & fare_condition

# --- 3. Apply the final condition/mask to the SCALED data (df_scaled) ---

# This filters the scaled data, keeping only the rows that met the unscaled IQR criteria
df_no_outliers = df_scaled[final_condition].reset_index(drop=True)

print(f"Original scaled rows: {len(df_scaled)}")
print(f"Rows after outlier removal (Age and Fare): {len(df_no_outliers)}")
print("\nHead of the final outlier-free, scaled DataFrame:")
print(df_no_outliers.head())
```

```python
# --- Assuming df_encoded (unscaled, encoded) and df_scaled (scaled, encoded) are available ---

# 1. Calculate IQR bounds using the UN-SCALED data (df_encoded)
Q1_age = df_encoded['Age'].quantile(0.25)
Q3_age = df_encoded['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age
lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age

Q1_fare = df_encoded['Fare'].quantile(0.25)
Q3_fare = df_encoded['Fare'].quantile(0.75)
IQR_fare = Q3_fare - Q1_fare
lower_bound_fare = Q1_fare - 1.5 * IQR_fare
upper_bound_fare = Q3_fare + 1.5 * IQR_fare

# 2. Create a boolean mask (conditions) based on the UN-SCALED data
age_condition = (df_encoded['Age'] >= lower_bound_age) & (df_encoded['Age'] <= upper_bound_age)
fare_condition = (df_encoded['Fare'] >= lower_bound_fare) & (df_encoded['Fare'] <= upper_bound_fare)

# Combine the conditions
final_condition = age_condition & fare_condition

# 3. Apply the final condition/mask to the SCALED data (df_scaled)
df_no_outliers = df_scaled[final_condition].reset_index(drop=True)

# Print results to show the effect
print(f"Original scaled rows: {len(df_scaled)}")
print(f"Rows after outlier removal: {len(df_no_outliers)}")
print("\nHead of the final outlier-free, scaled DataFrame:")
print(df_no_outliers.head())
```
